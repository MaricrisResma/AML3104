#### <b>Q1. Describe the decision tree classifier algorithm and how it works to make predictions.



- The decision tree classifier algorithm is a non-parametric supervised learning method that is used for classification problems. The algorithm works by splitting the data into smaller subsets based on the independent features one at a time until pure nodes are reached. Each node represents feature and the branches of the tree represents a decision.
- It consists of several components namely parent root node, decision nodes and  leaf nodes which come form the independent features.
- Evaluating the formation of each node in the tree is made by calculating the entropy, information gain, or gini impurity from the  probabilities of each category for classification.
- It can then be used to make predictions on which class something belongs to by following the flow of the tree, step by step in each decision for each node/feature.

#### <b>Q5. Define the confusion matrix and describe how it can be used to evaluate the performance of a classification model.

- A confusion matrix is an evaluation metric to check the performance of your classification model. It is a table with 4 different combinations of  predicted values and actual values for all the classes. It is a table that shows the number of true positives, true negatives, false positives and false negatives. True positives and true negatives are the number of predictions that were correct. On the other hand, false postives and false negatives are the number of wrong predictions.
- By having the number of correct predictions vs incorrect predictions, a confusion matrix can help measure percentage of the following:
  - accuracy - Out of all the data (TP+FP+TN+FN), how much have you predicted correctly as both true positive and true negative
  - precision - Out of all your predictions(of a specific class), how much was correct 
  - recall - Out of all the actual values(of a specific class), how much was correctly predicted
  - F1 score  - measures both precision and recall

#### <b>Q7. Discuss the importance of choosing an appropriate evaluation metric for a classification problem and explain how this can be done. </b>
- Understanding the cost of the error and weighing the consequences is what will help choose the appropriate evaluation metric.Things to weigh to understand the appropriate evaluation metric are as follows:
    - the class balance of data (accuracy will be biased to major class if hihgly imbalance therefore not a real representative of how well the model did)
    - the gravity of having wrong predictions for a certain class (if very important then precision should be high)
    (out of all the predicted class, is it ok that some are wrong by how much?)
    - the gravity of missing to some predictions for a specific classification (if important then recall should be high)
    (out of all the actual class, is it ok to miss some of it? by how much?)
    - both the gravity of having wrong predictions and gravity of missing to predict? (if important then f1 score should be high)
